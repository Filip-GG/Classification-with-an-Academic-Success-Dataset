```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import  OneHotEncoder
from sklearn.compose import ColumnTransformer

from sklearn.tree import DecisionTreeClassifier

import pandas as pd
import seaborn as sns
import catboost
```

Загрузим дата сет

```python
file_dir = './dataset/'
train_dit_name = 'train.csv'
test_dit_name = 'test.csv'
```

```python
train_set = pd.read_csv(file_dir+train_dit_name)
train_set.T
```

```python
train_set.info()
```

Расчитаем кореляцию и перобразуем дата сет

```python
plt.hist(train_set['Target'])
```

```python
train_set['Target'].value_counts()
```

```python
data_dict = {
    'Graduate': 0,
    'Dropout': 1 ,
    'Enrolled': 2
}
```

```python
def process_dataframe(df):
    df_copy = df.copy()
    df_copy['Target'] = list(map(lambda item: data_dict[item], df_copy['Target']))
    
    return df_copy
```

```python
train_dataset = process_dataframe(train_set)
train_dataset = train_dataset.drop(['id'], axis=1)
train_dataset.T
```

нормализуем данные:

```python
cor  = train_dataset.corr()

fig, ax = plt.subplots(figsize=(16,12)) 
sns.heatmap(cor, ax=ax)
```

Hапишем dataset

```python
dataset = train_dataset
```

```python
lable_train = dataset['Target'].to_numpy()

lable_train
```

```python
class_weight = {
    0:  36282/76518,
    1:  25296/76518,
    2:  14940/76518,
}
```

```python
dataset = dataset.drop(['Target'], axis=1)
data_train = dataset.to_numpy()
data_train
```

Модель

```python
model = catboost.CatBoostClassifier(
    iterations=21000,
    learning_rate=1e-2,
    loss_function='MultiClass',
    #class_weights = class_weight,
    auto_class_weights='SqrtBalanced',
    task_type="GPU",
    devices='0'
)

model
```

```python
model.fit(data_train, lable_train)
```

Skore

```python
model.score(data_train, lable_train)
```

Predickt

```python
test_data = pd.read_csv(file_dir+test_dit_name)
test_data = test_data.drop(['id'], axis=1)
data_test = test_data.to_numpy()
```

```python
out = model.predict(data_test)
```

```python
pd.DataFrame({
    'id':range(76518+0, 76518+len(data_test)),
    'Target':list(map(lambda item: ['Graduate', 'Dropout', 'Enrolled'][item], out.reshape(-1)))
}).to_csv('sample_submission.csv', index=False)
```
